**Descriptive Analysis**

```python
# 1. Kategori produk yang paling sering dibeli
product_count = df.groupby('Category')['Product ID'].count()
print(product_count)
```

```python
# Membuat diagram batang jumlah produk yang terjual di setiap kategori
plt.bar(product_count.index, product_count.values, color=['blue', 'red', 'green'])
plt.xticks(rotation=0)
plt.xlabel('Category')
plt.ylabel('Total Products Sold')
plt.title('Total Products Sold by Category')
plt.show()
```

Office supplies merupakan kategori produk yang paling sering dibeli sebanyak dari 5909 total keseluruhan

```python
# 2. Berapa jumlah total produk yang terjual di setiap negara bagian
product_count = df.groupby('State')['Product ID'].count()
print(product_count)
```

**Diagnostic Analytics**

```python
#1. Apakah ada peningkatan atau penurunan yang signifikan dalam ship mode selama periode 2017 berdasarkan ship date

# Mengganti kolom ship date menjadi tipe data datetime
#df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Filter data untuk periode tahun 2017
start_date = pd.to_datetime('2017-01-01')
end_date = pd.to_datetime('2017-12-31')
filtered_data = df[(df['Ship Date'] >= start_date) & (df['Ship Date'] <= end_date)]

# Menghitung jumlah pengiriman untuk setiap ship mode
ship_mode_counts = filtered_data['Ship Mode'].value_counts()
ship_mode_counts
```

```python
# Membuat visualisasi bar chart
plt.bar(ship_mode_counts.index, ship_mode_counts.values)
plt.xlabel('Ship Mode')
plt.ylabel('Jumlah Pengiriman')
plt.title('Jumlah Pengiriman berdasarkan Ship Mode (2017)')
plt.xticks(rotation=45)
plt.show()
```

Bar chart di samping menunjukkan bahwa Standard class lebih sering dipilih untuk pengiriman, tidak ada peningkatan dari ship mode yang lain

```python
#2. Berapa jumlah pesanan di setiap tahun pada dataset berdasrkan order date

# Mengganti kolom order date menjadi tipe data datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Mengambil tahun dari kolom order_date
df['Year'] = df['Order Date'].dt.year

# Menghitung jumlah pesanan berdasarkan tahun
order_counts_by_year = df['year'].value_counts().sort_index()

# Menampilkan hasil
print("Tahun", "Jumlah Pesanan")
print(order_counts_by_year)

# Membuat visualisasi grafik
plt.figure(figsize=(10, 6))
order_counts_by_year.plot(kind='bar', color='skyblue')
plt.title('Jumlah Pesanan per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Pesanan')
plt.xticks(rotation=45)
plt.show()
```

Jumlah pesanan meningkat disetiap tahunnya, tahun 2018 menempati posisi puncak dengan total pesanan sebanyak 3258 pesanan

```python
# 3. Apa saja segmen pelanggan Terbaik

segment_sales = df.groupby('Segment')['Sales'].sum().reset_index()

plt.figure(figsize=(8, 6))
plt.bar(segment_sales['Segment'], segment_sales['Sales'], color='skyblue')
plt.title('Total Penjualan per Segmen Pelanggan')
plt.xlabel('Segmen Pelanggan')
plt.ylabel('Total Penjualan')
plt.show()

best_segment = segment_sales[segment_sales['Sales'] == segment_sales['Sales'].max()]
print("Segmen Pelanggan Terbaik:", best_segment['Segment'].values[0])
```

```python
# 4. Analisis rentabilitas dari 10 produk yang ada

df_sorted = df.sort_values(by='Sales', ascending=False)

most_profitable_product = df_sorted.iloc[0]['Product Name']
print("Produk Paling Menguntungkan:", most_profitable_product)

least_profitable_product = df_sorted.iloc[-1]['Product Name']
print("Produk Paling Merugi:", least_profitable_product)

plt.figure(figsize=(8, 6))
plt.bar(df_sorted['Product Name'][:10], df_sorted['Sales'][:10], color='skyblue')
plt.title('10 Produk Paling Menguntungkan')
plt.xlabel('Produk')
plt.ylabel('Profit')
plt.xticks(rotation=90)
plt.show()
```
